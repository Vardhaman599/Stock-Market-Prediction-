```python
#Import necessary libraries.
```

```python
# sample code for data exploration
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Sample code for linear regression implementation
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Loading the Data set
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/NXTIVIA INTERNSHIP/Stock Market Dataset.csv")
```

```python
from google.colab import data_table
from vega_datasets import data
```

```python
#Data Preprocessing
```

```python
#select relevant features for prediction
features=['Tesla_Vol.']
#split the dataset into features (X) and target variable [y]
x=df[features]
y=df['Tesla_Price']
```

```python
#Train Test and split and linear Regression Model
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#initialize and train the linear regression model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#Make prediction as the test set
y_pred=model.predict(x_test)

#Evaluate the model
mse=mean_squared_error(y_test,y_pred)
print(f'mean square Error:{mse}')
```

```python
#use the trained model to predict Tesla_Price on  the set
y_pred_test= model.predict(x_test)

#compare the predicted values the actual values
test_results =pd.DataFrame({'Actual':y_test,'Predicted':y_pred_test})
print(test_results)

#visualize the comparison
plt.figure(figsize=(12,6))
plt.plot(test_results['Actual'],label= 'Actual Tesla_Price',marker='o')
plt.plot(test_results['Predicted'],label ='Predicted Tesla_Price',marker='x')
plt.title('Comparison of Actual vs Predicted Tesla_price')
plt.xlabel('sample Index')
plt.ylabel('Tesla_price(USD)')
plt.legend()
plt.show()
```
